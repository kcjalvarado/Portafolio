### Convertir columnas objeto/str a int

```python
df["Weight"] = df["Weight"].str.replace("lbs","")
```

```python
df["Weight"] = pd.to_numeric(df["Weight"])
```

```python
def convert_height_to_cm(height_str):
    hsm = height_str.replace("\"","")
    feet, inches = map(int, hsm.split("'"))
    total_inches = feet * 12 + inches
    centimeters = total_inches * 2.54
    return centimeters
```

```python
df["Height"] = df["Height"].apply(convert_height_to_cm)
df["Height"] = pd.to_numeric(df["Height"])
```

### Handmade monetary convertions

```python
def monetaryConvertion(incval):
    pivstr = incval.replace("â‚¬","")
    if "K" in pivstr:
        intstr = pivstr.replace("K","")
        retstr = float(intstr)*1000
    elif "M" in pivstr:
        intstr = pivstr.replace("M","")
        retstr = float(intstr)*1000000
    else:
        retstr = pivstr
    
    return int(retstr)
    
    
```

```python
df["Wage"] = df["Wage"].apply(monetaryConvertion)
```

```python
df["Release Clause"] = df["Release Clause"].apply(monetaryConvertion)
```

```python
df["Value"] = df["Value"].apply(monetaryConvertion)
```

```python
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1
outlier_indices = df.index[((df['Value'] < Q1 - 1.5 * IQR) | (df['Value'] > Q3 + 1.5 * IQR))]
```

```python
sbn.scatterplot(x="Wage", y="Value", data=df, hue=df.index)

# Overlay outlier points with distinct color and marker
sbn.scatterplot(x="Wage", y="Value", data=df.loc[outlier_indices], color="red", marker="x", s=150)
```

#### using BOKEH

```python
from bokeh.models import ColumnDataSource, HoverTool, NumeralTickFormatter
from bokeh.plotting import figure, show
```

```python
source = ColumnDataSource(df)
```

```python
p = figure(title="Wage vs. Value", x_axis_label="Wage", y_axis_label="Value")

# Plot circles with color mapped to position (adjust "position" if needed)
p.circle(x="Wage", y="Value", source=source, size=10, color="red")

# Hover tool displays player name and other details
hover = HoverTool()
hover.tooltips = [
    ("Player:", "@LongName"),
    ("Wage:", "@Wage{$0,0}"),
    ("Value:", "@Value{$0,0}"),
]
p.add_tools(hover)

# Format numeric axes
p.yaxis.formatter = NumeralTickFormatter(format="$0,0")
p.xaxis.formatter = NumeralTickFormatter(format="$0,0")

show(p)
```
